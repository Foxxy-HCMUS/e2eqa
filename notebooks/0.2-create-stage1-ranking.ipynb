```python
import json
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
trainval = json.load(open("../data/processed/zac2022_train_merged_final.json"))
```

```python
trainval
```

```python
raw_df = pd.DataFrame(trainval['data'])
full_anno = raw_df[raw_df['category'] == 'FULL_ANNOTATION']
train, val = train_test_split(full_anno, random_state=42, test_size=0.15)
full_anno
```

```python
df = raw_df.copy()
df['label'] = np.where(((raw_df['category'] == 'FULL_ANNOTATION') | (raw_df['is_long_answer'])), 1, 0)

df = df.loc[~df.index.isin(val['id'])]

df = df[['text', 'question', 'label']]
df['text'] = [x['dirty_text'] or x['text'] for x in trainval['data'] if x['id'] not in val['id']]
df['text'] = df['text'].str.replace('\r', '\n')
df
```

```python
df.to_csv("../data/processed/train_stage1_ranking.csv",index=False)
```
